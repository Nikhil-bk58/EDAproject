```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


df = pd.read_csv("/content/spotify_churn_dataset.csv")


print(df.head())
print(df.info())
```

```python
plt.figure()
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.hist(df['age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['listening_time'])
plt.title("Listening Time Boxplot")
plt.show()
```

```python
plt.figure()
plt.scatter(df['age'], df['listening_time'])
plt.title("Age vs Listening Time")
plt.xlabel("Age")
plt.ylabel("Listening Time")
plt.show()
```

```python
# Sorting by age for better visualization
df_sorted = df.sort_values(by='age')

plt.figure()
plt.plot(df_sorted['age'], df_sorted['songs_played_per_day'])
plt.title("Age vs Songs Played Per Day")
plt.xlabel("Age")
plt.ylabel("Songs Played")
plt.show()
```

```python
df['subscription_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Subscription Type Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x='subscription_type', hue='is_churned', data=df)
plt.title("Subscription vs Churn")
plt.show()
```

```python
fig = px.scatter(df, x='age', y='listening_time',
                 color='subscription_type',
                 title="Interactive Age vs Listening Time")
fig.show()
```

```python
fig = px.bar(df, x='country', title="Users by Country")
fig.show()
```

```python
fig = px.pie(df, names='subscription_type', title="Subscription Distribution")
fig.show()
```
